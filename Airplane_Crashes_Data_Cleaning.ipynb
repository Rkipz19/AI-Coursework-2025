Q5: Create a new dataframe from the raw/uncleaned data called
âfatality_locationsâ with the following columns: âDateâ,
âLocationâ, âAboardâ, âFatalitiesâ.

```python
fatality_locations = df[['Date','Location','Aboard','Fatalities']].copy()

fatality_locations.head()
```

Q6: Determine the date of the highest number of recorded fatalities from this new
dataframe.

```python
highest_fatality = fatality_locations.loc[fatality_locations['Fatalities'].idxmax()]

print("Date of the highest fatality recorded: ", highest_fatality['Date'])
```

```python
fatality_locations[['Aboard','Fatalities']].head()

noFatalities = fatality_locations[fatality_locations['Fatalities'] == 0]
NumberOfNoFatalities = noFatalities.shape[0]
print("Number of crashes with no fatalities: ", NumberOfNoFatalities)
```

Q8: In the âfatality_locationsâ dataframe, the location column has data items
represented as âRegion, U.S. State/Countryâ. Split this column to have two columns
of regions and states/countries

```python
fatality_locations[['Region', 'State/Country']] = fatality_locations['Location'].str.split(',', n=1, expand=True )
fatality_locations[['Location', 'Region', 'State/Country']].head()
```

Q9: Order the dataframe by the number of fatalities from highest to lowest and select the
first 100.

```python
top100Fatalities = fatality_locations.sort_values(by='Fatalities', ascending=False).head(100)
top100Fatalities.head()
```

Q10: Generate a pie chart that shows the distribution of the top 25 number of fatalities
per country/U.S. State.

```python
import seaborn as sns
import matplotlib.pyplot as plt

top25 = top100Fatalities.head(25)
top25 = fatality_locations.sort_values(by='Fatalities',ascending=False).head(25)
fatalitiesByCountry = top25.groupby('State/Country')['Fatalities'].sum().sort_values(ascending=False)
plt.figure(figsize=(32,32))
plt.pie(fatalitiesByCountry, labels=fatalitiesByCountry.index, autopct='1%.1f%%', startangle=140)
plt.title('Distribution of top 25 fatalities per country/U.S State')
plt.show()
```
